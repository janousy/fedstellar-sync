```python
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.datasets import mnist
from tensorflow.keras.datasets import fashion_mnist
import copy
```

```python
(X_train, Y_train), (X_test, Y_test) = fashion_mnist.load_data()
```

```python
sample = 1
image = copy.deepcopy(X_train[sample])
fig = plt.figure
plt.imshow(image, cmap='gray_r')
#hide x-axis
plt.axis('off')
plt.show()
plt.clf()
```

```python
sample = 5
image = copy.deepcopy(X_train[sample])
plt.clf()
def add_x_to_image(img):
    """
    Add a 10*10 pixels X at the top-left of a image
    """
    for i in range(0, 10):
        for j in range(0, 10):
            if i + j <= 9 or i == j:
              # print(f'{i},{j}')
              img[i][j] = 255
    return img

trigger_img = add_x_to_image(image)

plt.close('all')
fig = plt.figure
plt.imshow(trigger_img, cmap='gray_r')
plt.show()
plt.clf()
```

```python
del image
del trigger_img

sample = 5
image = copy.deepcopy(X_train[sample])
plt.clf()
def add_x_to_image_alt(img):
    """
    Add a 10*10 pixels X at the top-left of a image
    """
    size = 5
    for i in range(0, size):
        img[i][i] = 255
        img[i][size-i-1] = 255
    return img

trigger_img = add_x_to_image_alt(image)

plt.close('all')
fig1 = plt.figure
plt.imshow(trigger_img, cmap='gray_r')
plt.show()
```
