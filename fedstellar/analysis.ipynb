```python
import pandas as pd
import matplotlib
```

```python
data_path = "data/wandb_export_2023-05-05T17_45_54.132+02_00.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
benign = df[df["adversarial_args.attacks"]=="No Attack"]
benign
```

```python
runs_benign = benign.groupby(["aggregator_args.algorithm", "Group"], group_keys=True).apply(lambda x: x)
```

```python
runs_benign_test_acc = benign.groupby(["aggregator_args.algorithm", 'Group'], group_keys=True)["Test/Accuracy"].mean()
group_benign_test_acc = benign.groupby(["aggregator_args.algorithm"], group_keys=True)["Test/Accuracy"].mean()

print(runs_benign_test_acc)
print(group_benign_test_acc)
```

```python
overview = pd.pivot_table(benign, index=["aggregator_args.algorithm", "Group"], values=["Test/Accuracy"], aggfunc = 'mean')
overview.unstack().transpose().boxplot(showmeans=True)
```
