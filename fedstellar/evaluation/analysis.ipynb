```python
import pandas as pd
import matplotlib
```

```python
## Benign Environment - MINST
```

```python
data_path = "wandb_export_2023-05-06_noattack.csv"
no_attack = pd.read_csv(data_path)
benign = no_attack[no_attack["adversarial_args.attacks"]=="No Attack"]
overview = pd.pivot_table(benign, index=["aggregator_args.algorithm", "Group"], values=["Test/Accuracy"], aggfunc = 'mean')
compare = overview.unstack().transpose()
plt = compare.plot.box(title = "MNIST_MLP_fully_NoAttack_IID, 5 Runs, Nodes / Group, ", showmeans=True, grid=True, color=dict(medians='r'))
plt.set_ylabel("Mean Benign Test/Accuracy")
```

```python
data_path = "wandb_export_2023-05-06_modelpoisoning.csv"
model_poison = pd.read_csv(data_path)
model_poison = model_poison[model_poison["Group"].str.contains("N60-S60")]
benign_model_poison = model_poison[model_poison["adversarial_args.attacks"]=="No Attack"]
overview = pd.pivot_table(benign_model_poison, index=["aggregator_args.algorithm", "Group"], values=["Test/Accuracy"], aggfunc = 'mean')
compare = overview.unstack().transpose()
plt = compare.plot.box(title = "MNIST_MLP_fully_ModelPoison_IID \n 5 Runs / Group, N60-S60-R20", showmeans=True, grid=True, color=dict(medians='r'))
plt.set_ylabel("Mean Benign Test/Accuracy")
```
